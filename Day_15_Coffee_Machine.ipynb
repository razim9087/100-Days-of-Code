```python
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}

status=resources
status["Money"]=0
uom=["ml","ml","g","$"]
units=dict(zip(status.keys(),uom))
```

```python
def print_report(status,units):
  for i in status.keys():
    if i!="Money":
      print(f"{i}: {status[i]}{units[i]}")
    else:
      print(f"{i}: {units[i]}{status[i]}")

def availability_check(prompt,status,MENU):
  w=MENU[prompt]["ingredients"]
  k=False
  for i in w.keys():
    if status[i]<=w[i]:
      k+=True
      break
    else:
      pass
  if k:
    print(f"Not enough {i}")
    return False
  else:
    return True

def calculate_entry(a):
  print("Please insert money!!")
  q=int(input("Enter Quarters:"))
  d=int(input("Enter Dimes:"))
  n=int(input("Enter Nickles:"))
  p=int(input("Enter Pennies:"))
  total=q*0.25+d*0.1+n*0.05+p*0.01
  change=total-a
  if change>0:
    message="Transaction successfull.Please collect the change."
  elif change==0:
    message="Transaction successfull."
  else:
    message="Transaction unsuccessfull.Money refunded."
  return total,message

def status_update(prompt,status):
  for i in MENU[prompt]["ingredients"].keys():
    status[i]-=MENU[prompt]["ingredients"][i]
  return status

def refill(resources,status):
  for j in resources.keys():
    status[j]=resources[j]

machine_state="On"
while machine_state=="On":
  prompt=input("What would you like? (expresso/latte/cappuccino)")
  if prompt=="Off":
    machine_state="Off"
    refill(resources,status)
    break
  elif prompt=="Report":
    print_report(status,units)
  elif prompt in ["expresso","latte","cappuccino"]:
    e=availability_check(prompt,status,MENU)
    if e:
      paid,message=calculate_entry(MENU[prompt]["cost"])
      if paid>=MENU[prompt]["cost"]:
        status["Money"]+=MENU[prompt]["cost"]
        status=status_update(prompt,status)
        print(message)
        print(f"Please collect your {prompt}. Visit again!!")
      else:
        print(message)
        break
    else:
      break

```
