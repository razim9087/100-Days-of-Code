```python
import random

def hit(deal,cards):
  deal.append(cards[random.randint(0,len(cards)-1)])
  return deal

def blackjack_game(computer_deal,player_deal,cards):
  decision=False
  print(f"Computer's first card: {computer_deal[0]}")
  your_total=player_deal[0]+player_deal[1]
  print(f"Your cards: {player_deal}, Current score: {your_total}")
  while sum(player_deal)<21:
    dc=input("Type ""h"" to get another card or ""s"" to stand:")
    if dc=="h":
      player_deal=hit(player_deal,cards)
      if sum(player_deal)>21:
        print(f"Your hand of {player_deal} exceeds 21. You Lose!!")
        decision=True
        break
      else:
        print(f"Your hand is {player_deal} with total {sum(player_deal)}")
        continue
    elif dc=="s":
      break
    else:
      continue
  computer_total=sum(computer_deal)
  while computer_total<17 and sum(player_deal)<=21:
    computer_deal=hit(computer_deal,cards)
    if sum(computer_deal)==21 and sum(player_deal)<21:
      print(f"Computer has blackjack with the hand {computer_deal}. You Lose.")
      decision=True
      break
    elif sum(computer_deal)==21 and sum(player_deal)==21:
      print("Both have blackjacks. Its a draw!!")
      decision=True
      break
    elif sum(computer_deal)>21:
      print(f"Computer's hand of {computer_deal} exceeds 21. You Win!!")
      decision=True
      break
    else:
      computer_total=sum(computer_deal)
      continue
  if not decision:
    if sum(player_deal)>sum(computer_deal):
      print(f"You win with hand {player_deal} angainst computer's hand {computer_deal}.")
    elif sum(player_deal)==sum(computer_deal):
      print(f"Its a draw with your hand {player_deal} angainst computer's hand {computer_deal}.")
    else:
      print(f"You lose with hand {player_deal} angainst computer's hand {computer_deal}.")
```

```python
play=input("Do you want to play a game of Blackjack? Type ""y"" or ""n"":")
cards=[11,2,3,4,5,6,7,8,9,10,10,10,10]
computer_deal=[]
player_deal=[]
if play=="y":
  for i in range(2):
    computer_deal.append(cards[random.randint(0,len(cards)-1)])
    player_deal.append(cards[random.randint(0,len(cards)-1)])
  blackjack_game(computer_deal,player_deal,cards)
else:
  print("Have a good day!!")
```
