```python
print("Welcome to the Secret Auction!!")
maximum_bid=int(input("Enter maximum bid:"))
bids={}
winning_bids={}
take_bid=True
starting_bid=int(input("Enter starting bid:"))

def record_bid(take_bid):
  process=input("Are there any more bidders? Enter Yes or No:")
  if process=="No":
    take_bid=False
    name=0
    bid=-1
    return name,bid,take_bid
  else:
    name=input("Enter your name:")
    bid=int(input("Enter your bid:"))
    return name,bid,take_bid

def aggregate_bid(take_bid,bids):
  while take_bid:
    a,b,take_bid=record_bid(take_bid)
    if b<=maximum_bid and (b>=starting_bid or b==-1):
      bids[a]=b
    else:
      print("Invalid Bid. Bid again!!")
  return bids

def bid_comparison(winning_bids,bids):
  bids=aggregate_bid(take_bid,bids)
  max=0
  max_count=0
  winning_bids[list(bids.keys())[0]]=list(bids.values())[0]
  check=True if len(list(bids.values()))>1 else False
  if check:
    for i in range(1,len(list(bids.values()))):
      if list(bids.values())[i]>list(bids.values())[i-1]:
        del winning_bids[list(bids.keys())[i-1]]
        winning_bids[list(bids.keys())[i]]=list(bids.values())[i]
      elif list(bids.values())[i]==list(bids.values())[i-1]:
        winning_bids[list(bids.keys())[i]]=list(bids.values())[i]
      else:
        pass
    return winning_bids
  else:
    return winning_bids

while len(winning_bids.keys())!=1:
  print(f"Welcome to a new round of bidding!\nBidding starts at {starting_bid}")
  take_bid=True
  winning_bids=bid_comparison(winning_bids,bids)
  starting_bid=list(winning_bids.values())[0]

print(f"Bidding closes at {list(winning_bids.values())[0]}.\nThe winner of the bidding process is {list(winning_bids.keys())[0]}.")
```
