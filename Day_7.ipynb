```python
from wonderwords import RandomWord

r = RandomWord()
# Generate a single random word
word = r.word()
print(f"Generated word: {word}")
```

```python
def input_check(a):
  b=input("Enter Letter:")
  if a!=b:
    return False
  else:
    return True


print(f"You have to guess a {len(word)}-letter word.")
life=""
placeholder=""
i=0
j=0
while life!="hangman":
  if input_check(word[i]):
    placeholder=placeholder+word[i]
    i=i+1
    x1=len("hangman")-len(life)
    y1=len(word)-len(placeholder)
    print(f"You have {x1} chances left to guess {y1} letters.")
    if placeholder==word:
      print("You Won!!")
      break
    elif y1>x1:
      print("You Lose!")
      break
    elif y1<=x1:
      print("Correct Guess.Proceed to next letter.")
  else:
    life=life+"hangman"[j]
    j=j+1
    x2=len("hangman")-len(life)
    y2=len(word)-len(placeholder)
    print(f"You have {x2} chances left to guess {y2} letters.")
    if life=="hangman" or y2>x2:
      print("You Lose!!")
      break
    else:
      print("Guess again.")
```
