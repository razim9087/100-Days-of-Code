```python
import random
print("Welcome to Rock-Paper-Scissors!!")
result=0
while result==0:
  your_turn=int(input("Enter 0 for rock, 1 for paper, 2 for scissor: "))
  if your_turn==0:
     print("You chose Rock!")
  elif your_turn==1:
     print("You chose Paper!")
  else:
     print("You chose Scissor!")
  computer_turn=random.randint(0,2)
  if computer_turn==0:
     print("Computer chose Rock!")
  elif computer_turn==1:
     print("Computer chose Paper!")
  else:
     print("Computer chose Scissor!")
  if your_turn==computer_turn:
    print("Play again !!")
  else:
    result=1
    if your_turn==0 and computer_turn==2:
      print("You Win!!")
    elif your_turn==1 and computer_turn==2:
      print("You Lose!!")
    elif your_turn==0 and computer_turn==1:
      print("You Lose!!")
    elif your_turn==2 and computer_turn==1:
      print("You Win!!")
    elif your_turn==1 and computer_turn==0:
      print("You Win!!")
    elif your_turn==2 and computer_turn==0:
      print("You Lose!!")
```
