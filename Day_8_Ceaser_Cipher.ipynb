```python
alphabet=[]
for i in range(97,123):
  alphabet.append(chr(i))
#alphabet
```

```python
direction=input("Enter ""encode"" to encrypt or ""decode"" to decrypt:").lower()
text=input("Enter text:").lower()
amount=int(input("Enter shift number:"))

def ceaser_cipher(direction,text,amount,key):
  if direction=="encode":
    output=""
    for k in text:
      if k in key:
        m=key.index(k)
        n=m+amount
        p=n%len(key)
        output=output+key[p]
      else:
        output=output+k
    return output
  else:
    output=""
    for k in text:
      if k in key:
        m=key.index(k)
        n=m-amount
        p=(n+len(key))%len(key)
        output=output+key[p]
      else:
        output=output+k
    return output

ceaser_cipher(direction,text,amount,alphabet)
```
