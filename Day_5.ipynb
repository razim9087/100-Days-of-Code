```python
import random
print("Welcome to PyPassword Generator!")

capital_letters=int(input("Enter number of capital letters in your password:"))
small_letters=int(input("Enter number of small letters in your password:"))
digits=int(input("Enter number of digits in your password:"))
special_characters=int(input("Enter number of special characters in your password:"))
password=""
input_list=[capital_letters,small_letters,digits,special_characters]
output_list=[0,0,0,0]
total_length=sum(input_list)
i=0
while i<total_length:
  a1=random.randint(0,3)
  a2=input_list[a1]
  a3=output_list[a1]
  if a3<a2:
    a3+=1
    i+=1
    if a1==0:
      password=password+chr(random.randint(65,90))
    elif a1==1:
      password=password+chr(random.randint(97,122))
    elif a1==2:
      password=password+chr(random.randint(0,9))
    elif a1==3:
      password=password+chr(random.randint(48,64))
print(password)
```
