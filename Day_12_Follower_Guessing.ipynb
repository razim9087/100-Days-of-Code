```python
import random

print("Welcome to Follower Guessing Game!!")
i=random.randint(0,len(data)-1)
j=[]
j.append(i)
def random_selector(j):
    is_random = False
    while not is_random:
        k=random.randint(0,len(data)-1)
        if k in j:
            pass
        else:
            is_random = True
            j.append(k)
    return k,j

defeat=False
exhausted=False
while not defeat and not exhausted:
    print("Who has more followers?")
    print(f"Option A : {data[i]['name']}, {data[i]['description']} from {data[i]['country']}")
    print("\n")
    g,j=random_selector(j)
    print(f"Option B: {data[g]['name']}, {data[g]['description']} from {data[g]['country']}")
    response=input("Enter your choice, A or B:")
    if response=="A":
        user_choice,compare=i,g
    else:
        user_choice, compare = g,i
    if data[user_choice]['follower_count']>=data[compare]['follower_count']:
        if len(j)<len(data):
            i=user_choice
        else:
            print("You Win!!")
            exhausted=True
    else:
        print("You Lose!!")
        defeat=True
```
