```python
import random
print("Welcome to the number guessing game!!")
difficulty=input("Enter ""easy"" for easy mode or ""difficult"" for difficult mode:")
turns=0
if difficulty=="easy":
  turns=10
elif difficulty=="difficult":
  turns=5
else:
  turns=5
guess=random.randint(1,100)
for i in range(turns):
  y=int(input("Enter your guess:"))
  if y==guess:
    print("You guessed it!!")
    break
  elif y-guess>20 and i!=turns-1:
    print("Too high. Guess again.")
  elif guess-y>20 and i!=turns-1:
    print("Too low. Guess again.")
  elif y-guess>=10 and y-guess<=20 and i!=turns-1:
    print("High. Guess again.")
  elif guess-y>=10 and guess-y<=20 and i!=turns-1:
    print("Low. Guess again.")
  elif y-guess>=5 and y-guess<10 and i!=turns-1:
    print("Close but high. Guess again.")
  elif guess-y>=5 and guess-y<10 and i!=turns-1:
    print("Close but low. Guess again.")
  elif guess!=y and i==turns-1:
    print("Out of turns. You Lose!!")
  else:
    print("Almost there. Guess carefully.")
```
