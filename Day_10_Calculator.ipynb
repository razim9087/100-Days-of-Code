```python
continue_calc=True
operand=False
def add(f,s):
  return f+s

def subtract(f,s):
  return f-s

def multiply(f,s):
    return f*s

def divide(f,s):
    return f/s

def exponent(f,s):
    return f**s

dict={"+":add,"-":subtract,"*":multiply,"/":divide,"^":exponent}

while continue_calc:
  if not operand:
    a=int(input("Enter first number:"))
    operand=True
  else:
    b=int(input("Enter second number:"))
    c=input("Available operations:\n+\n-\n*\n/\n^\nSpecify operation:")
    while c not in dict.keys():
      c=input("Invalid Operand. Enter operator again.")
    a=dict[c](a,b)
    print(f"The result is {a}.")
    d=input(f"Press y to continue with {a} and n to terminate session.")
    if d=="n":
      continue_calc=False
    else:
      pass
```
