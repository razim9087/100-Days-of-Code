```python
continue_calc=True
operand=False
def calculator(f,s,op):
  if op=="+":
    return f+s
  elif op=="-":
    return f-s
  elif op=="*":
    return f*s
  elif op=="/":
    return f/s
  elif op=="^":
    return f**s
  else:
    return "Unidentified Operator!!"

while continue_calc:
  if not operand:
    a=int(input("Enter first number:"))
    operand=True
  else:
    b=int(input("Enter second number:"))
    c=input("Available operations:\n+\n-\n*\n/\n^\nSpecify operation:")
    result=calculator(a,b,c)
    if type(result)==type("str"):
      while type(result)==type("str"):
        e=input("Invalid Operand. Enter operator again.")
        result=calculator(a,b,e)
        a=result
    else:
      a=result
    print(f"The result is {result}.")
    d=input(f"Press y to continue with {result} and n to terminate session.")
    if d=="n":
      continue_calc=False
    else:
      pass
```
