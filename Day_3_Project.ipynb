```python
print("Welcome to Treaure Island.\nYour mission is to find the treasure.")
first_move=input("Which direction are you heading? Press L for Left and R for Right.")
if first_move=="L":
  second_move=input("Will you wait or swim? Press W to wait and S to swim.")
  if second_move=="W":
    third_move=input("You have three doors to choose from. Press R to enter Red door,B to enter Blue door, Y to enter Yellow door")
    if third_move=="Y":
      print("You Win!!!")
    elif third_move=="R":
      print("You were burned by fire and died. Game Over!")
    elif third_move=="B":
      print("You were eaten by beasts and died. Game Over!")
    else:
      print("Game Over!")
  else:
    print("You got attacked by trout and died. Game Over!")
else:
  print("You fell into a hole and died. Game Over!")
```
